```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import sem

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
Combined_Data = pd.merge(mouse_metadata,study_results,on="Mouse ID", how="left")
# Display the data table for preview
Combined_Data
```

```python
# Checking the number of mice.
Mouse_count = len(Combined_Data["Mouse ID"].unique())
Mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
Duplicate_Mouse = Combined_Data.loc[Combined_Data.duplicated(["Mouse ID","Timepoint"])]
Duplicate_Mouse_id = Duplicate_Mouse["Mouse ID"].unique()
Duplicate_Mouse_id
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#Clean_df = Combined_Data.drop(Duplicate_Mouse.index)
Clean_df = Combined_Data[Combined_Data["Mouse ID"].isin(Duplicate_Mouse_id)==False]
Clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
Clean_Mouse_count = len(Clean_df["Mouse ID"].unique())
Clean_Mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
Table_data = Clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"]
mean = Table_data.mean()
median = Table_data.median()
variance = Table_data.var(ddof=0) 
standard_deviation = Table_data.std(ddof=0)
sem_value = Table_data.sem()
Summary_Dataframe =pd.DataFrame({
                                  "Mean Tumor Volume": mean,
                                  "Median Tumor Volume":median,
                                  "Tumor Volume Variance":variance,
                                  "Tumor Volume Std. Dev.":standard_deviation,
                                  "Tumor Volume Std. Err.":sem_value
                                 })
Summary_Dataframe
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
Advanced_Summary = Clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(mean="mean",median="median",var="var",std="std",sem="sem")
Advanced_Summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
Treatment_Regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
Filter_Regimens = Clean_df.loc[Clean_df["Drug Regimen"].isin(Treatment_Regimens)]
Final_Tumor = Filter_Regimens.groupby("Mouse ID")["Tumor Volume (mm3)"].last()

# Start by getting the last (greatest) timepoint for each mouse
Last_timepoint = Filter_Regimens.groupby("Mouse ID")["Timepoint"].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
Merge_df = pd.merge(Last_timepoint,Clean_df,on=["Mouse ID","Timepoint"],how="left")
Merge_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
Treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
# Create empty list to fill with tumor vol data (for plotting)
Tumour_Vol_List = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for Treatment in Treatments:
    Final_tumour_volumes = Merge_df[Merge_df["Drug Regimen"]==Treatment]
    # Locate the rows which contain mice on each drug and get the tumor volumes
    Tumor_vol = Final_tumour_volumes.loc[Final_tumour_volumes["Drug Regimen"]==Treatment,["Tumor Volume (mm3)"]]
    Tumour_Vol_List.append(Tumor_vol)

    # add subset 
    Quartiles = Tumor_vol["Tumor Volume (mm3)"].quantile([.25,.5,.75])
    Lowerq = Quartiles[0.25]
    Upperq = Quartiles[0.75]
    IQR = Upperq-Lowerq
    Lower_Bound = Lowerq - (1.5*IQR)
    Upper_Bound = Upperq + (1.5*IQR)

    # Determine outliers using upper and lower bounds
    Outliers = Tumor_vol.loc[(Tumor_vol["Tumor Volume (mm3)"] > Upper_Bound)|(Tumor_vol["Tumor Volume (mm3)"] < Lower_Bound)]
    print(f"{Treatment} 's potential outliers: {Outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
Treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
Plot_Data = []
for Treatment in Treatments:
    Final_tumour_volumes = Merge_df[Merge_df["Drug Regimen"]==Treatment]
    Final_Tumor_Vol = Final_tumour_volumes.loc[Final_tumour_volumes["Drug Regimen"]==Treatment,"Tumor Volume (mm3)"]
    Plot_Data.append(Final_Tumor_Vol)
ax1.set_ylabel("Final Tumor Volume (mm3)")
ax1.boxplot(Plot_Data,labels= Treatments,flierprops=dict(marker='o', markerfacecolor='r', markersize=10))
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
Capomulin_Data = Clean_df.loc[Clean_df["Drug Regimen"]=="Capomulin"]
#Capomulin_Data["Mouse ID"].unique()
Capomulin_Data_L509 = Capomulin_Data.loc[Capomulin_Data["Mouse ID"]=="l509"]
plt.plot(Capomulin_Data_L509["Timepoint"],Capomulin_Data_L509["Tumor Volume (mm3)"])
plt.title("Capomulin Treatment of Mouse L509")
plt.xlabel("Timepoint(days)")
plt.ylabel("Tumor Volume (mm3)")

plt.show
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
Average_Vol = Capomulin_Data.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()
Mouse_Weight = Capomulin_Data.groupby("Mouse ID")["Weight (g)"].mean()
Scatter_Data = pd.merge(Average_Vol,Mouse_Weight,on="Mouse ID",how="left")
Scatter_Data.plot(kind="scatter", x="Weight (g)", y="Tumor Volume (mm3)",figsize=(6,4))
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
Correlation = np.round(st.pearsonr(Scatter_Data["Weight (g)"], Scatter_Data["Tumor Volume (mm3)"])[0],2)
print(f'The correlation between mouse weight and the average tumor volume is {Correlation}')

slope, intercept, r_value, p_value, std_err = st.linregress(Scatter_Data["Weight (g)"],Scatter_Data["Tumor Volume (mm3)"])
fit_line = slope * Scatter_Data["Weight (g)"]+ intercept

Scatter_Data.plot(kind="scatter", x="Weight (g)", y="Tumor Volume (mm3)",figsize=(6,4))
plt.plot(Scatter_Data["Weight (g)"],fit_line, color="r")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
```
